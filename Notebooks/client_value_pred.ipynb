### Changed a few dimensions in the data_input.csv

```python
df_input = pd.read_csv('data_input.csv')
```

```python
df_input.head()
```

```python
df_input = df_input.drop('Unnamed: 0', axis=1)
```

```python
df_input = df_input.assign(client_service=df_input["client"] + "_" + df_input["service"])
```

```python
df_input = df_input.set_index("client_service")
```

```python
df_input = df_input.drop("client",axis=1)
```

```python
df_input
```

### Encoding the service column

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['service']
le = LabelEncoder()

df_input[categorical_cols] = df_input[categorical_cols].apply(lambda col: le.fit_transform(col))
```

### Test and train data split

```python
x = df_input.drop('transactions_cy', axis=1)
y = df_input['transactions_cy']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 28)
```

### Model fitting and training

```python
model = RandomForestRegressor(n_estimators = 10, random_state = 0)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

rmse = float(format(np.sqrt(mean_squared_error(y_test, y_pred)), '.3f'))
print("\nRMSE: ", rmse)
```
