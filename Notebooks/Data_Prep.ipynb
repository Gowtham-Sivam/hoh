### Subscriptions.csv

```python
import csv
import random

seed_value = 47
random.seed(seed_value)

year_range = (2010, 2022)
roi_range = (1, 100)
feedback_range = (1, 5)
num_services_range = (1, 10)

min_cost_per_unit = 50  
max_cost_per_unit = 100 

clients_data = []
with open('clients.csv', mode='r') as clients_file:
    clients_reader = csv.DictReader(clients_file)
    for row in clients_reader:
        clients_data.append(row)

services_data = []
with open('services.csv', mode='r') as services_file:
    services_reader = csv.DictReader(services_file)
    for row in services_reader:
        services_data.append(row)


with open('subscription.csv', mode='w', newline='') as subscription_file:
    fieldnames = ['client', 'year', 'service', 'roi', 'cost', 'feedback']
    writer = csv.DictWriter(subscription_file, fieldnames=fieldnames)
    writer.writeheader()


    for client in clients_data:
        prev_year_cost = random.uniform(min_cost_per_unit, max_cost_per_unit)

        for year in range(year_range[0], year_range[1] + 1):
            num_subscriptions = random.randint(num_services_range[0], num_services_range[1])

            for _ in range(num_subscriptions):
                service = random.choice(services_data)
                roi = random.randint(roi_range[0], roi_range[1])

               
                cost_adjustment = random.uniform(-60, 60)  

                # Apply the adjustment factor to the previous year's cost
                cost = max(min(prev_year_cost + cost_adjustment, max_cost_per_unit), min_cost_per_unit)

                feedback = random.randint(feedback_range[0], feedback_range[1])

                subscription = {
                    'client': client['client'],
                    'year': year,
                    'service': service['service'],
                    'roi': roi,
                    'cost': cost,
                    'feedback': feedback
                }

                writer.writerow(subscription)

            # Update the previous year's cost for the next iteration
            prev_year_cost = cost
```

### client_activity.csv

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta


random.seed(42)


Transaction_Type_List = [
    "Trade Execution",
    "Cash Settlement",
    "Reconciliation",
    "Contract Termination",
    "Onboarding",
    "Contract Renewal",
    "FX Transaction",
    "Corporate Action",
    "Inquiry"
]


start_year = 2019
end_year = 2023

# Create a function to generate random dates within the specified range
def generate_random_date(start_year, end_year):
    start_date = datetime(start_year, 1, 1)
    end_date = datetime(end_year, 12, 31)
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    return random_date.strftime('%Y-%m-%d')

# Create a function to generate client activity data
def generate_client_activity(interpolated_subscription, start_year, end_year, min_amount, max_amount):
    client_activity_data = []

    for _, row in interpolated_subscription.iterrows():
        client = row['client']
        service = row['service']
        transaction_type = random.choice(Transaction_Type_List)
        activity_date = generate_random_date(start_year, end_year)
        amount = 0 if transaction_type == "Inquiry" else round(random.randint(min_amount, max_amount), 2)

        client_activity_data.append([client, service, transaction_type, activity_date, amount])

    return pd.DataFrame(client_activity_data, columns=['client', 'service', 'transaction_type', 'activity_date', 'amount'])

# Load the interpolated_subscription data from CSV
interpolated_subscription = pd.read_csv('interpolated_subcriptions.csv')

# Generate client activity data
client_activity = generate_client_activity(interpolated_subscription, start_year, end_year, 10, 1000)

# Save the client_activity data to CSV
client_activity.to_csv('client_activity.csv', index=False)
client_activity
```

```python
import pandas as pd
import random

# Configurable parameters
roi_percentage_range = (0, 50)  # Define your ROI percentage range here
year_range = (2020, 2022)  # Define your year range here
min_services_per_client = 1  # Define the minimum number of services per client
max_services_per_client = 10  # Define the maximum number of services per client

# Read client CSV
client_df = pd.read_csv('C:/Users/hoh_user/Desktop/Hackathon_2023/Dataset/V2_Datasets/Client.csv')

# Read service CSV
service_df = pd.read_csv('C:/Users/hoh_user/Desktop/Hackathon_2023/Dataset/V2_Datasets/Services.csv')

# Initialize an empty subscription dataframe
subscription_df = pd.DataFrame(columns=['Client id', 'Service id', 'Cost', 'Year', 'ROI', 'Feedback'])

# Generate subscription data
for _, client_row in client_df.iterrows():
    client_id = client_row['CLIENTID']
    
    # Determine the random number of services for this client
    num_services = random.randint(min_services_per_client, max_services_per_client)
    
    # Randomly sample 'num_services' rows from the service dataframe
    sampled_services = service_df.sample(num_services)
    
    for _, service_row in sampled_services.iterrows():
        service_id = service_row['SERVICE ID']
        service_cost = service_row['COST OF SERVICE']
        
        year = random.randint(year_range[0], year_range[1])
        roi = random.uniform(roi_percentage_range[0], roi_percentage_range[1])
        feedback = random.randint(1, 5)
        
        subscription_df = pd.concat([subscription_df, pd.DataFrame({
            'Client id': [client_id],
            'Service id': [service_id],
            'Cost': [service_cost],
            'Year': [year],
            'ROI': [roi],
            'Feedback': [feedback]
        })], ignore_index=True)

# Save the subscription data to a CSV file
subscription_df.to_csv('C:/Users/hoh_user/Desktop/Hackathon_2023/Dataset/V2_Datasets/subscription_res.csv', index=False)
print("Subscription data has been generated and saved to subscription.csv.")
```
